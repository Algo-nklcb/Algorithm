```python
from collections import deque

directions = [(0, 1), (-1, 0), (1, 0), (0, -1)]
def bfs(arr, x, y, shark_size):
    n = len(arr)
    ans = []
    dist = [[-1]*n for _ in range(n)]
    q = deque()
    q.append((x, y))
    dist[x][y] = 0
    while q:
        v = q.popleft()
        for dx, dy in directions:
            nx, ny = dx + v[0], dy + v[1]
            if 0 <= nx < n and 0 <= ny < n:
                if dist[nx][ny] == -1: # 지나가지 않은 곳
                    ok = False # 갈 수 있는지 여부
                    eat = False # 먹을 수 있는지 여부
                    
                    if arr[nx][ny] == 0: # 빈 공간 지나감
                        ok = True
                    elif arr[nx][ny] < shark_size: # 상어보다 작으므로 먹고 지나갈 수 있음
                        ok = True
                        eat = True
                    elif arr[nx][ny] == shark_size: # 크기가 같으므로 먹진 못하고 지나갈 수 있음
                        ok = True
                        
                    if ok: # 지나갈 수 있으면
                        q.append((nx, ny))
                        dist[nx][ny] = dist[v[0]][v[1]] + 1
                        if eat:
                            ans.append((dist[nx][ny], nx, ny)) # 물고기까지의 거리, 그 물고기의 x좌표, y좌표
    if not ans:
        return None
    
    ans.sort() # 정렬하게 되면 거리순, 윗쪽 순, 왼쪽 순으로 정렬됨.
    return ans[0] # 가장 가깝고, 위에 있고, 왼쪽에 있는 것이 return됨                 


N = int(input())
arr = []
for _ in range(N):
    arr.append(list(map(int, input().split())))


start_x, start_y = 0, 0
for i in range(N):
    for j in range(N):
        if arr[i][j] == 9:
            start_x, start_y = i, j
            arr[i][j] = 0 # 아기상어 위치 저장 후 0으로 업데이트

            
shark_size = 2 # 상어 크기 
eat_count = 0 # 먹은 개수
ans = 0 # 엄마 상어에게 도움을 요청하지 않고 물고기를 잡아먹을 수 있는 시간
while True:
    p = bfs(arr, start_x, start_y, shark_size)
    if p is None:
        break
    dist, nx, ny = p # 먹은 물고기까지의 거리, 물고기 x좌표, 물고기 y좌표
    arr[nx][ny] = 0 # 먹었으니 0으로 업데이트
    ans += dist # 물고기 먹기 위한 그 시간을 더해줌
    eat_count += 1 # 먹은 개수 1 증가
    if eat_count == shark_size:
        shark_size += 1
        eat_count = 0
    start_x, start_y = nx, ny
    
print(ans)
```
